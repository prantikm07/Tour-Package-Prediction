```python
df_c = pd.get_dummies(df,drop_first = True)
```

```python
df_c.shape
```

```python
from sklearn.impute import KNNImputer
df1 = pd.DataFrame(KNNImputer().fit_transform(df_c),columns=df_c.columns)
```

```python
df1.isnull().sum()
```

```python
x = df1.drop('ProdTaken',axis=1)
y = df1['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred  = model.predict(x_test)
acc = accuracy_score(y_test,y_pred)
acc
```

```python
#Feature selection
ft = model.feature_importances_
imp = pd.Series(ft,index=x.columns).sort_values()
plt.figure(figsize=(5,10))
imp.plot.barh()
plt.show()
```
